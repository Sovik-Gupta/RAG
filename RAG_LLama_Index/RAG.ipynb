```python
from llama_index.core import VectorStoreIndex,SimpleDirectoryReader,ServiceContext
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core.prompts.prompts import SimpleInputPrompt
```

```python
documents = SimpleDirectoryReader("/content/drive/MyDrive/Colab Notebooks/LLM/").load_data()
len(documents)
```

```python
documents=documents[0:30]
```

```python
system_prompt = """
you are a QA assistant. your goal is to answer question as accuratlt as possible based on
the instructions and context provided
"""
```

```python
# Default prompt suuported by llama2

query_wrapper_prompt = SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
!!huggingface-cli login
```

```python
llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",
    device_map="auto"
    # model_kwargs={"torch_dtype":torch.float16, "load_in_8bit":True}
)
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index.core import ServiceContext
from llama_index.legacy.embeddings.langchain import LangchainEmbedding
# from llama_index.embeddings.langchain import LangchainEmbedding


embed_model = LangchainEmbedding(HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2"))
```

```python
service_context = ServiceContext.from_defaults(
    chunk_size=1024,
    llm=llm,
    embed_model=embed_model)
```

```python
index=VectorStoreIndex.from_documents(documents, service_context=service_context)
```

```python
query_engine=index.as_query_engine()
```

```python
response=query_engine.query("what is Hypothesis Testing")
```
